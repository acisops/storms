```python
from astropy.io import ascii
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
fns = list(Path(".").glob("*.ecsv"))
```

```python
fig, ax = plt.subplots(figsize=(10,10))
for fn in fns:
    t = ascii.read(fn)
    ax.loglog(t["p3"], t["hrc_shield"], 'x', color="C0")
    idx = None
    for cmt in t.meta["comments"]:
        if "idx" in cmt:
            print('shutdown')
            idx = int(cmt.split()[-1][:-1])
            break
    if idx is not None:
        row = t[idx]
        ax.plot(row["p3"], row["hrc_shield"], 'x', color="C3", mew=5, ms=10, zorder=100)
```
